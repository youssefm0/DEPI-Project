```python
## Highest profitable items
```

```python
item_revenue = df.groupby('StockItem_Name')['Revenue'].sum()
```

```python
top_10_items = item_revenue.nlargest(10).index.tolist()
```

```python
df['Is_Top_10'] = df['StockItem_Name'].apply(lambda x: x if x in top_10_items else None)
```

```python
df_filtered = df[df['Is_Top_10'].notna()]
```

```python

sns.barplot(data=df_filtered, x='Lead_Time_Days', y='Revenue', hue='Is_Top_10', alpha=0.6)
plt.title('Revenue vs. Lead Time (Days) for Top 10 Stock Items')
plt.show()
```
